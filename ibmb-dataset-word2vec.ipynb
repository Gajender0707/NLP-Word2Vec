```python
import re
```

```python
# Downloads the data.
import nltk
nltk.download('stopwords')


# Using the stopwords.
from nltk.corpus import stopwords

# Initialize the stopwords
stoplist = stopwords.words('english')
```

```python
def preprocessing(x):
    x=x.lower()
    x=re.sub("[^a-zA-Z]"," ",x)
    x=x.split()
    x=[lemmatizer.lemmatize(word) for word in x if word not in stoplist]
    x=" ".join(x)
    return x
```
